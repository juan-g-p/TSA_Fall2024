```python
import os
import pandas as pd
from fuzzywuzzy import process
import unicodedata
from openpyxl import load_workbook
```

```python
def read_named_range(workbook, range_name):
    named_range = workbook.defined_names[range_name]
    values = []

    for title, coord in named_range.destinations:  # may include multiple worksheets
        ws = workbook[title]
        for row in ws[coord]:
            for cell in row:
                values.append(cell.value)
    return values

def normalize_string(s):
    """Normalize strings by lowercasing and stripping spaces."""
    return s.lower().strip()
```

```python
# Load the workbook
excel_path = 'ZZ_TSA_G1_midterm.xlsm'  # Update this to your Excel file path
workbook = load_workbook(filename=excel_path, data_only=True)
```

```python
# Read data from named ranges
names = read_named_range(workbook, 'names')
surnames = read_named_range(workbook, 'surnames')

# Ensure names and surnames list are of the same size
if len(names) != len(surnames):
    raise ValueError("Names and surnames lists are of different sizes.")

# Combine names and surnames
students = [f"{n} {s}" for n, s in zip(names, surnames)]
```

```python
# List all files in the directory
folder_path = '.'  # Update this to your folder path
files = os.listdir(folder_path)

# Create a mapping from normalized names to original names
norm_to_orig = {normalize_string(student): student for student in students}

# Process each file
for file in files:
    normalized_file = normalize_string(file)
    best_match, score = process.extractOne(normalized_file, list(norm_to_orig.keys()))
    if score >= 90:  # Adjust the threshold as needed
        original_name = norm_to_orig.get(best_match)
        if original_name:
            student_files[original_name].append(file)

# Results
for student, matched_files in student_files.items():
    print(f"Student: {student}, Files: {matched_files}")
```

```python
import os

def normalize_string(s):
    """Normalize strings by lowercasing."""
    return s.lower()

# Assume 'names' and 'surnames' are already read from the Excel file
# ...

# Keep original and normalized student names together
students = [{"original": f"{n} {s}", "normalized": normalize_string(f"{n} {s}")} for n, s in zip(names, surnames)]

# List all files in the directory
folder_path = '.'  # Update with your folder path
files = os.listdir(folder_path)

# Initialize a dictionary to hold the mapping of students to their files
student_files = {student["original"]: [] for student in students}

# Process each file
for file in files:
    normalized_file = normalize_string(file)
    for student in students:
        if student["normalized"] in normalized_file:
            student_files[student["original"]].append(file)

# Display the results
for student, matched_files in student_files.items():
    print(f"Student: {student}, Files: {matched_files}")
```
